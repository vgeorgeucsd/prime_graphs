```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
n = 10
p = 0.3
G = nx.gnp_random_graph(n, p, seed=None, directed=True)
```

```python
H = nx.Graph()
for e in G.edges():
    src=str(e[0])
    tar=str(e[1])
    prime=str(e[1])+'p'
    H.add_edge(src,prime)
    H.add_edge(prime,tar)
```

```python
pos = nx.shell_layout(G)
nx.draw_networkx_nodes(G,pos=pos)
nx.draw_networkx_edges(G,pos=pos)
nx.draw_networkx_labels(G,pos=pos)
```

```python
top = nx.bipartite.sets(H)[0]
pos = nx.bipartite_layout(H,top)
nx.draw_networkx_nodes(H,pos=pos)
nx.draw_networkx_edges(H,pos=pos)
nx.draw_networkx_labels(H,pos=pos)
```

```python
DG=nx.DiGraph()
import re

for e in H.edges():
    if(re.search('[a-zA-Z]', e[0])):
        src = e[0].replace('p','')
        tar = e[1]
        if(src==tar):
            continue
        DG.add_edge(tar,src)
    if(re.search('[a-zA-Z]', e[1])):
        src = e[0]
        tar = e[1].replace('p', '')
        if(src==tar):
            continue
        DG.add_edge(src,tar)

```

```python
print(nx.is_isomorphic(G,DG))
```
